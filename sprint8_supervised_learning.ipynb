## Improve the quality of the model. Make sure you use at least two approaches to fixing class imbalance. Use the training set to pick the best parameters. Train different models on training and validation sets. Find the best one. Briefly describe your findings.

````python
# Approach 1: Upsampling
def upsample(features, target, repeat):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_upsampled = pd.concat([features_zeros, features_ones] + [features_ones] * (repeat - 1))
    target_upsampled = pd.concat([target_zeros, target_ones] + [target_ones] * (repeat - 1))
    features_upsampled, target_upsampled = sklearn.utils.shuffle(features_upsampled, target_upsampled, random_state=12345)
    return features_upsampled, target_upsampled

features_upsampled, target_upsampled = upsample(features_train, target_train, 4)
# Identify categorical columns (replace with your actual categorical column names)
categorical_cols = ['Geography', 'Gender']

# One-hot encode the upsampled training data
features_upsampled_encoded = pd.get_dummies(features_upsampled, columns=categorical_cols, drop_first=True)

# One-hot encode the validation data
features_valid_encoded = pd.get_dummies(features_valid, columns=categorical_cols, drop_first=True)

# Align columns between training and validation sets
features_upsampled_encoded, features_valid_encoded = features_upsampled_encoded.align(features_valid_encoded, join='left', axis=1, fill_value=0)
````

````python
# Approach 2: Downsampling
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]
    features_downsampled = pd.concat([features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat([target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    features_downsampled, target_downsampled = sklearn.utils.shuffle(features_downsampled, target_downsampled, random_state=12345)
    return features_downsampled, target_downsampled

features_downsampled, target_downsampled = downsample(features_train, target_train, 0.25)

# Identify categorical columns (replace with your actual categorical column names)
categorical_cols = ['Geography', 'Gender']

# One-hot encode the downsampled training data
features_downsampled_encoded = pd.get_dummies(features_downsampled, columns=categorical_cols, drop_first=True)

# One-hot encode the validation data
features_valid_encoded = pd.get_dummies(features_valid, columns=categorical_cols, drop_first=True)

# Align columns between training and validation sets
features_downsampled_encoded, features_valid_encoded = features_downsampled_encoded.align(features_valid_encoded, join='left', axis=1, fill_value=0)
````

````python
# Train models with upsampled data
pipeline_upsampled_lr = Pipeline(steps=[('preprocessor', preprocessor),
                                        ('classifier', LogisticRegression(random_state=12345, solver='liblinear'))])
pipeline_upsampled_rf = Pipeline(steps=[('preprocessor', preprocessor),
                                        ('classifier', RandomForestClassifier(random_state=12345))])

pipeline_upsampled_lr.fit(features_upsampled, target_upsampled)
pipeline_upsampled_rf.fit(features_upsampled, target_upsampled)

target_valid_pred_upsampled_lr = pipeline_upsampled_lr.predict(features_valid)
target_valid_pred_upsampled_rf = pipeline_upsampled_rf.predict(features_valid)

f1_upsampled_lr = f1_score(target_valid, target_valid_pred_upsampled_lr)
auc_roc_upsampled_lr = roc_auc_score(target_valid, pipeline_upsampled_lr.predict_proba(features_valid)[:, 1])

f1_upsampled_rf = f1_score(target_valid, target_valid_pred_upsampled_rf)
auc_roc_upsampled_rf = roc_auc_score(target_valid, pipeline_upsampled_rf.predict_proba(features_valid)[:, 1])

print(f"\nUpsampled LR F1: {f1_upsampled_lr}")
print(f"Upsampled LR AUC-ROC: {auc_roc_upsampled_lr}")
print(f"Upsampled RF F1: {f1_upsampled_rf}")
print(f"Upsampled RF AUC-ROC: {auc_roc_upsampled_rf}")
````
